

# 📂 Google Colab에서 Google Drive 마운트와 경로 설정하기

---

## 1. 🔍 구글 드라이브 마운트란?

**마운트(Mount)**란,  
Google Colab에서 **내 Google Drive를 연결하는 것**을 말합니다.  
쉽게 말해, 내 드라이브를 Colab 안에서 **폴더처럼 사용할 수 있게 만드는 과정**이에요.

> Colab은 클라우드에서 실행되기 때문에  
> 내 컴퓨터나 구글 드라이브에 있는 파일을 바로 사용할 수 없습니다.  
> 그래서 마운트 작업이 필요합니다.

---

## 2. ❓ 구글 드라이브 마운트는 왜 필요할까?

우리가 앞으로 작성할 코드에서는  
CSV, 이미지, 텍스트 파일 등 다양한 데이터를 다루게 됩니다.

이 데이터를 **매번 업로드하는 대신**,  
내 Google Drive에 저장해두고 **Colab에서 직접 불러올 수 있다면 훨씬 편리**하겠죠?

### ✅ 마운트의 장점
- 내 파일을 드라이브에만 올려두면 언제든지 불러올 수 있음
- 매번 파일을 업로드할 필요가 없음
- Colab과 드라이브 간 파일 저장/읽기 자유로움

---

## 3. 📌 구글 드라이브 마운트하는 방법

아래 코드를 Colab에 입력하고 실행하세요:

```python
from google.colab import drive
drive.mount('/content/drive')
```


이 코드는 왼쪽 파일 아이콘에서 드라이브마운트 아이콘을 클릭하면 자동 생성됩니다

## 4. 📁 마운트 후 경로 설정하기 (`%cd` 사용)

Google Drive를 마운트하면 기본적으로 Colab에서 다음 경로에 연결됩니다:


> /content/drive/MyDrive/



하지만 매번 이 경로를 전부 쓰기 번거롭기 때문에,  
작업할 폴더로 **작업 디렉토리(current directory)** 를 바꿔주는 것이 좋습니다.

이를 위해 `%cd` 명령어를 사용합니다.

<br/>
<br/>

---

### 4-1. 📍 내 파일 경로 찾기

1. Colab 왼쪽 메뉴에서 **폴더 아이콘(📁)** 을 클릭합니다.  
2. `drive` → `MyDrive` 로 들어가면 내 Google Drive의 폴더들이 보입니다.  
3. 내가 작업할 폴더를 찾아 들어갑니다.  
4. 원하는 폴더나 파일에서 **오른쪽 클릭 > 경로 복사** 를 선택하세요.

> 예시:  
> `/content/drive/MyDrive/data`
